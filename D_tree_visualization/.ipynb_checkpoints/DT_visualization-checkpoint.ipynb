```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier 
from sklearn import tree
import pandas as pd
from sklearn.model_selection import cross_val_score , train_test_split
from sklearn.metrics import confusion_matrix , plot_confusion_matrix
```

```python
# Prepare the data data
df = pd.read_csv("data/Iris.csv")
df.head()
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
sns.pairplot(df,hue = "Species")
```

Iris-setosa is linearly seperable but other two species features are non-linearly seperable, so we need KNN or Decison tree to make classification

```python
# Preprocessing
X = df.drop(columns = ["Id","Species"])
y = df.Species
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size = 0.2, random_state = 1234)

# Fit the classifier with default hyper-parameters
clf = DecisionTreeClassifier(random_state=1234)
clf.fit(X_train,y_train)
```

```python
# Confusion matrix
plot_confusion_matrix(clf,X_test,y_test)
```

```python
scores = cross_val_score(clf, X, y, cv=5, scoring='f1_macro')
print(f"Average F1 score: {scores.mean()} with standard deviation of {scores.std()}")
```

```python
text_representation = tree.export_text(clf)
print(text_representation)
```

```python
fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(clf, 
                   feature_names=X.columns,  
                   class_names=df.Species,
                   filled=True)
```
